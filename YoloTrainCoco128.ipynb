**Pasos a seguir para entrenar un modelo Detect**

>1¬∫ Importamos YOLO para tener acceso a los metodos necesarios

>2¬∫Seguimos seleccioanndo el tipo del modelo que queremos usar, en este caso el m√°s sencillo ya que es el que mejor se adapta a todas las necesidades

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
model.train(data='coco128.yaml', epochs=2) 
```

```python
result = model(source="prueba1.jpg", show=True, conf=0.4, save=True)
```

***Modelo pose***
El modelo Pose se encarga de la estimaci√≥n de posici√≥n, que es la tarea de encontrar puntos espec√≠ficos en una imagen, com√∫nmente conocidos como puntos clave. Los puntos clave pueden representar varias partes de un objeto, como conexiones, puntos de referencia u otras caracter√≠sticas. Las ubicaciones de los puntos clave generalmente se expresan como un conjunto de coordenadas 2D [x, y] o 3D [x, y, visible].

Como resultado obtenemos los puntos clave de un objeto en la imagen, generalmente junto con las puntuaciones de confianza para cada punto. La estimaci√≥n de pose es una buena opci√≥n cuando necesitas identificar partes espec√≠ficas de un objeto en una escena y su ubicaci√≥n entre s√≠.

```python
# Load YOLOv8n-pose, train it on COCO8-pose for 3 epochs and predict an image with it
from ultralytics import YOLO

model = YOLO('yolov8n-pose.pt')  # load a pretrained YOLOv8n classification model
model.train(data='coco8-pose.yaml', epochs=3)  # train the model
```

```python
result = model(source=0, show=True, conf=0.4, save=True)
```

```python
result = model(source="pruebapose.jpg", show=True, conf=0.4, save=True)
```
